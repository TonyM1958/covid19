```python
import covid
```

```python
# United Kingdom
# lag and dilation adjusted so prediction is similar to actual
covid.Region(geoId='UK', lag=2, dilation=1).analyse(ylog=1)
```

```python
# China
covid.Region(geoId='CN', dilation=1).analyse(ylog=1)
```

```python
# Spain
covid.Region(geoId='ES', dilation=1.8).analyse()
```

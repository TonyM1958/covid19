<br>Example region codes: UK, FR, IT, DE, ES, US, CN, KR, JP, RU, NZ, AU, BE, IE


```python
import covid
```

```python
# United Kingdom
# lag and dilation set so prediction tracks latest data
covid.Region(geoId='UK', dilation=2.5, d_cases=1).analyse(ylog=1)
```

```python
# China
covid.Region(geoId='CN', dilation=1).analyse(ylog=1)
```

```python
# Spain
covid.Region(geoId='ES', dilation=3.3).analyse()
```

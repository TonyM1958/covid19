Get the latest json data from <a target="_blank" href="https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide">www.ecdc.europa.eu</a>

<br>Example region codes: UK, FR, IT, DE, ES, US, CN, KR, JP, RU, NZ, AU, BE


```python
import covid
```

```python
# United Kingdom
# lag and dilation set so prediction tracks latest data
covid.Region(geoId='UK', dilation=0.75, d_cases=2).analyse(ylog=1)
```

```python
# China
covid.Region(geoId='CN', dilation=1).analyse(ylog=1)
```

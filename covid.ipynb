```python
import covid
ylog=1
```

```python
# United Kingdom
covid.Region(geoId='UK', dilation=2.1, d_cases=1.6).analyse(ylog=ylog)
```

```python
# China
covid.Region(geoId='CN', dilation=1).analyse(ylog=1)
```
